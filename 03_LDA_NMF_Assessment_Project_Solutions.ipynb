# Preprocessing

#### Task: Use TF-IDF Vectorization to create a vectorized document term matrix. You may want to explore the max_df and min_df parameters.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

```python
dtm = tfidf.fit_transform(uj['Cleaned_text'])
```

```python
dtm
```

# Non-negative Matrix Factorization

#### TASK: Using Scikit-Learn create an instance of NMF with 20 expected components. (Use random_state=42)..

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(n_components=20,random_state=42)
```

```python
nmf_model.fit(dtm)
```

```python
tfidf.get_feature_names_out()[1500]
```

#### NAO RODA - TASK: Print our the top 15 most common words for each of the 20 topics.


```python
for index,topic in enumerate(nmf_model.components_):
    print(f'THE TOP 10 WORDS FOR TOPIC #{index}')
    print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

#### TASK: Add a new column to the original quora Cleaned_textframe that labels each question into one of the 20 topic categories.

```python
uj.head()
```

```python
topic_results = nmf_model.transform(dtm)
```

```python
topic_results.argmax(axis=1)

uj['Topic'] = topic_results.argmax(axis=1)

uj.head(10)
```
